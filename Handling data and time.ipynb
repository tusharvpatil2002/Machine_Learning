# working with dates

```python
date.head()
```

```python
date = date.drop(columns=['product_id','city_id','orders'],axis=1)
```

```python
date.head()
```

```python
# converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

## extract year

```python
date.head()
```

```python
date['year'] = date['date'].dt.year
```

```python
date.head()
```

## extract month

```python
# month number
date['month'] = date['date'].dt.month

# month name
date['month_name'] = date['date'].dt.month_name()
```

```python
date.head()
```

## extract day

```python
date['day'] = date['date'].dt.day
```

```python
date.head()
```

```python
# which day of week
date['day_of_week'] = date['date'].dt.dayofweek
```

```python
date.head()
```

```python
# name of day
date['day_name'] = date['date'].dt.day_name()
```

```python
date.head()
```

## is weekend? if yes-> 1 otherwise 0

```python
date['is_weekend'] = np.where(date['day_name'].isin(['Sunday','Saturday']),1,0)
```

```python
date.head()
```

## extract week of year


```python
date['week_of_year'] = date['date'].dt.week
```

```python
date.head()
```

## extract quarter

```python
date['quarter'] = date['date'].dt.quarter
```

```python
date.head()
```

## extract semister

```python
date['semister'] = np.where(date['quarter'].isin([1,2]),1,2)
```

```python
date.head()
```
